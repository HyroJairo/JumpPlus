# 1. Load the csv file and show the top 5 records from it

```python
fifa = pd.read_csv('players_20.csv', index_col=[0])
fifa.head(5)
```

# 4. Show the number of players in each country


```python
player_per_country = fifa.groupby('nationality')
player_per_country.size()
```

# 5. Show the top 10 countries with the highest amount of players

```python
fifa_country_players = fifa.groupby('nationality')
fifa_country_players = fifa_country_players.size().nlargest(10)
fifa_country_players
```

# 6. Create a bar plot of the top 5 countries with the highest number of   players, and fill the bars in green

```python
import matplotlib.pyplot as plt
fifa_country_players = fifa.groupby('nationality')
top5 = fifa_country_players.size().nlargest(5)
top5 = top5.to_dict()
fig = plt.figure(figsize = (10, 5))
plt.bar(list(top5.keys()), list(top5.values()), color='green', width=0.4)
plt.xlabel('countries'), plt.ylabel('number of players')
plt.title('Top 5 countries with the highest number of players')
plt.show()
```

# 7. Display the short name and wages for the top 5 players


```python
fifa_shortname_wages_top5 = fifa[['short_name', 'wage_eur']]
fifa_shortname_wages_top5.head(5)
```

# 8. Display the short name and wages for the top 5 players with the highest salaries 

```python
top5 = fifa[['short_name', 'wage_eur']].sort_values('wage_eur', ascending=False).head(5)
top5
```

# 9. Create a bar plot of number 8

```python
top5 = pd.DataFrame.from_dict(top5)
top5 = dict(zip(top5['short_name'], top5['wage_eur']))
fig = plt.figure(figsize = (10,5))
plt.bar(list(top5.keys()), list(top5.values()), color='green', width=0.4)
plt.xlabel('players')
plt.ylabel('wage')
plt.title('Top 5 players who have the highest salary')
plt.show()
```
