# 1. Import required libraries


```python
import pandas as pd
```

# 2. Read the csv

```python
characters = pd.read_csv("charcters_stats.csv")
characters.dropna()
```

# 3. Show first record from csv

```python
# Used the head method to get the top row
characters.head(1)
```

# 4. Show number of rows and columns

```python
# Accessed the shape attribute to get the number of rows and columns
print(f"There are {characters.shape[0]} rows and {characters.shape[1]} columns")
```

# 5. You need to find the values of alignment

```python
# Used the value_counts() method shows the number of counts in each distinct value that is under one column
characters['Alignment'].value_counts()
```

# 6. Find out only good alignment holders superheroes

```python
# Accessing the dataframe through brackets
superheroes = characters[characters['Alignment'] == 'good']
superheroes
```

# 7. Show the first five records from point 6

```python
# Used head to find the first five records
superheroes.head(5)
```

# 8. Show the top five records having top speed of heroes of good alignment

```python
# Used the sort_values() method to sort the 'speed' column and ascending from high value to low value and get top 5 rows
superheroes.sort_values('Speed', ascending=False).head(5)
```

# 9. Show the 5 records of super heroes who have maximum power of good alignment

```python
# Used the sort_value to sort through the 'Power' column and scending from high to low and get top 5 rows
superheroes.sort_values('Power', ascending=False).head(5)
```

# 10. Find out how many super heroes are there with power 100 of good alignment

```python
# Created the dataframe of heroes with power equal to 100
Powerfuls = superheroes[superheroes['Power'] == 100]
print(f'There are {len(Powerfuls)} super heroes that have power of 100')
```

# 11. Shape them what you got in point 10

```python
# Accessed shape to find it's attributes
print(f"There are {Powerfuls.shape[0]} rows and {Powerfuls.shape[1]} columns")
```

# 12. Show all records from point 10

```python
# Just entered the dataframe to see all records
Powerfuls
```

# 13. Retrieve the first five records with max power and has good alignment

```python
# Used the query method to filter the alignment and power
characters.query("Alignment == 'good' and Power == 100").head(5)
```

# 14. Draw a bar plot of all super heroes who are having good alignment and max power of top five only, take the same object of point 13, show the name and total in plot with green bars

```python
# imported the matplotlib library and used figure method to setup the sizes and used bar() to place names and xlabel,xticks, and ylabel to give it the appropriate labels
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (20, 10))

plt.bar(Powerfuls['Name'], Powerfuls['Total'], color='green')
plt.xlabel('Superheroes', fontweight='bold', fontsize=15)
plt.xticks(rotation=60)

plt.ylabel('Total', fontweight='bold', fontsize=15)
```

# 15. Extract villians having bad alignment

```python
# Accessing the dataframe through brackets
villains = characters[characters['Alignment'] == 'bad']
```

# 16. Show the first five records of point 15

```python
# Accessed the first five rows
villains.head(5)
```

# 17. Show the top five fastest super villains in term of super speed

```python
# Sorted the villians df to find the speed column and sorting from highest value to lowest value with first five rows
villains.sort_values('Speed', ascending=False).head(5)
```

# 18. Top five super villains in terms of intelligence

```python
# Sorted the villians df to find the intelligence from high value to lower value to find first five rows
villains.sort_values('Intelligence', ascending=False).head(5)
```

# 19. Show who is most dangerous super villain after calculating their total (top 5 only)

```python
# Sorted the villains df by 'Total' column from high value to low value to find first five rows
villains.sort_values('Total', ascending=False).head(5)
```

# 20. Draw a histogram for speed of super heroes having size 10,5 provide speed in histogram for only good alignment super heroes, title should be "distribution of speed", xlabel should be "speed"

```python
# Used the figure, hist method to setup and used xlabel, ylabel, and title to put the approprate labels on
plt.figure(figsize=(10,5))
plt.hist(superheroes['Speed'], color='blue', edgecolor='white')
plt.xlabel('speed')
plt.ylabel('number of superheroes')
plt.title('distribution of speed')
plt.show()
```

# 21. Draw a histogram for combat of super villains having fig size 10,5 provide combat in histogram for only bad alignment super heroes, title='distribution of combat', xlabel='combat'

```python
# Used the figure, hist method to setup and used xlabel, ylabel, and title to put the approprate labels on
plt.figure(figsize=(10,5))
plt.hist(superheroes['Combat'], color='blue', edgecolor='black')
plt.xlabel('combat')
plt.ylabel('number of superheroes')
plt.title('distribution of combat')
plt.show()
```

# 22. Create a piechart of the percentage of superheros, villains, and neutrals

```python
# Created the label set and percentage set. Then I used those to inserted into the pie method to achieve the piechart
label = characters['Alignment'].dropna().unique()
percentage = characters['Alignment'].value_counts()
plt.pie(percentage, labels=label, autopct='%.0f%%')
```
